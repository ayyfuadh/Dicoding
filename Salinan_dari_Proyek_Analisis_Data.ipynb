## Visualization & Explanatory Analysis

Pertanyaan 1: Bagaimana data order setiap tahunnya?

```python
yearly_orders_df = all_df.resample(rule='Y', on='order_approved_at').agg({
    'order_id': 'nunique',
    'price': 'sum'
})

yearly_orders_df.index = yearly_orders_df.index.year
yearly_orders_df = yearly_orders_df.reset_index()

yearly_orders_df.rename(
    columns={
        'order_id': 'orders_count',
        'price': 'revenue'
    }, inplace=True
)

yearly_orders_df
```

```python
plt.figure(figsize=(10,5))
plt.plot(yearly_orders_df['order_approved_at'], yearly_orders_df['orders_count'], marker='o', linewidth=2, color='#72BCD4')
plt.title('Banyak Order per Tahun', loc='center', fontsize=20)
plt.xticks(fontsize=10, rotation=30)
plt.yticks(fontsize=10)
plt.show()
```

### Pertanyaan 2: Kategori Produk apa yang paling banyak dan paling sedikit terjual di tahun 2016?

```python
from tabulate import tabulate

# Filter data untuk tahun 2016
filtered_df = all_df[all_df['order_purchase_timestamp'].dt.year == 2016]

# Menghitung jumlah produk yang terjual berdasarkan kategori produk
sum_order_item_df = filtered_df.groupby('product_category_name_english').size().reset_index(name='order_count')

# Mengurutkan DataFrame berdasarkan jumlah pesanan (order_count) secara menurun
sum_order_item_df = sum_order_item_df.sort_values(by='order_count', ascending=False)

# Menampilkan produk yang paling banyak terjual
most_sold_product = sum_order_item_df.iloc[0]

# Menampilkan produk yang paling sedikit terjual
least_sold_product = sum_order_item_df.iloc[-1]

# Konversi ke format tabel
most_least_sold_table = [
    ["Produk yang paling banyak terjual di tahun 2016", most_sold_product['product_category_name_english'], most_sold_product['order_count']],
    ["Produk yang paling sedikit terjual di tahun 2016", least_sold_product['product_category_name_english'], least_sold_product['order_count']]
]

# Menampilkan tabel dengan garis
print(tabulate(most_least_sold_table, headers=["", "Kategori Produk", "Jumlah Pesanan"], tablefmt='pretty'))


```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

# Bar sebelah kiri
sns.barplot(x='order_count', y='product_category_name_english', data=sum_order_item_df.head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Penjualan Terbanyak", loc="center", fontsize=15)
ax[0].tick_params(axis='y', labelsize=12)

# Bar sebelah kanan
sns.barplot(x='order_count', y='product_category_name_english', data=sum_order_item_df.sort_values(by='order_count', ascending=True).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Penjualan Terkecil", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)

plt.suptitle("Kategori Produk apa yang paling banyak dan paling sedikit terjual di tahun 2016", fontsize=20)
plt.show()
```
